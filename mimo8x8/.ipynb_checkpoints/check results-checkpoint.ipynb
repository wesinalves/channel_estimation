```python
import numpy as np
import pandas as pd
from scipy.io import savemat, loadmat
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('all_nmse_hq_dense_-11_Nr8_Nt8_3GHz.txt', sep=" ", header=None, names = np.arange(-21, 22, 3))
```

```python
dataset
```

```python
dataset.loc[0,:].plot(style='--', label='avg')
dataset.loc[1,:].plot(style='s', label='min')
dataset.loc[2,:].plot(style='o', label='max')
plt.legend()
```

```python
dataset3Ghz = pd.read_csv('all_nmse_hq_dense_-11_Nr8_Nt8_3GHz.txt', sep=" ", header=None, names = np.arange(-21, 22, 3))
dataset5Ghz = pd.read_csv('all_nmse_hq_dense_-11_Nr8_Nt8_5GHz.txt', sep=" ", header=None, names = np.arange(-21, 22, 3))
dataset60Ghz = pd.read_csv('all_nmse_hq_dense_-11_Nr8_Nt8_60GHz.txt', sep=" ", header=None, names = np.arange(-21, 22, 3))
dataset60Ghz_mob = pd.read_csv('all_nmse_hq_dense_-11_Nr8_Nt8_60GHz_mob.txt', sep=" ", header=None, names = np.arange(-21, 22, 3))
randomized = pd.read_csv('all_nmse_hq_dense_256samples_-11_Nr8_Nt8_randomized.txt', sep=" ", header=None, names = np.arange(-21, 22, 3))
```

```python
dataset3Ghz.loc[0,:].plot(style='--', label='2.8GHz')
dataset5Ghz.loc[0,:].plot(style='--', label='5GHz')
dataset60Ghz.loc[0,:].plot(style='--', label='60GHz')
dataset60Ghz_mob.loc[0,:].plot(style='--', label='60GHz mob')
randomized.loc[0,:].plot(style='--', label='Random sparse')
plt.xlabel("SNRdB")
plt.ylabel("nmse avg")
plt.title("1-bit MIMO with histeresys - Dense Architecture")
plt.legend()
```

```python
num_samples = [10, 20, 50, 60, 100, 150, 200, 220, 240]
for s in num_samples:
    x = pd.read_csv(f'all_nmse_hq_dense_{s}samples_-11_Nr8_Nt8_60GHz.txt', sep=" ", header=None, names = np.arange(-21, 22, 3))
    x.loc[0,:].plot(style='--', label= str(s) + 'samples')

x = pd.read_csv('all_nmse_hq_dense_-11_Nr8_Nt8_60GHz.txt', sep=" ", header=None, names = np.arange(-21, 22, 3))
x.loc[0,:].plot(style='--', label='256 samples')

plt.xlabel("SNRdB")
plt.ylabel("nmse avg")
plt.title("1-bit MIMO 60GHz with hist - Dense Architecture")
plt.legend()    
```
